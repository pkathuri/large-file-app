**Data Understanding**


```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset into google colab
df=pd.read_csv('/content/drive/MyDrive/Train.csv')
```

Question 1:Which countries did their nationalities bring in the highest amount of revenue?

```python
# Listing the top 10 highest spending countries in terms of the total amount that their tourists spent
highest_spenders=df.groupby('country')['total_cost'].sum().astype('int').sort_values(ascending=False)
top_10_highest_spenders=highest_spenders.reset_index().head(10)
top_10_highest_spenders
```

```python
# A barplot of the top 10 countries in terms of total expenditure by tourists
plt.figure(figsize=(8,4),dpi=150)
plt.ticklabel_format(style='plain')
sns.barplot(x='country',y='total_cost',data=top_10_highest_spenders)
plt.xticks(rotation=90)
plt.title('Total expenditure by country.')
plt.gca().set(frame_on=False)
```

The chart above shows the top 10 countries whose nationationalities contributed the highest amount to the economy of Tanzania.The chart shows that nationalities if wealthy western nations and South Africa brought in the highest amount of revenue to the economy of Tanzania.

**Question 2:Why was the highest amount of money spent?**

```python
# Dataframe for the purpose of visiting Tanzania and rhe total money that was spent
visit_expenditure=df.groupby('purpose')['total_cost'].sum().astype('int').sort_values(ascending=False).reset_index()
visit_expenditure
```

```python
# A barplot of the top 10 countries in terms of total expenditure by tourists
plt.figure(figsize=(8,4),dpi=150)
plt.ticklabel_format(style='plain')
sns.barplot(x='purpose',y='total_cost',data=visit_expenditure)
plt.title('Total expenditure by purpose of visit.')
plt.xticks(rotation=90)
plt.gca().set(frame_on=False)
```

The chart above shows why the tourists who made a tour to Tanzania and the total amount of revenue they generated to the economy based on their purpose of visit to Tanzania. It is vivid that Leisure and Holidays generated the highest amount of revenues to the economy of Tanzania while other reasons followed distantly. 

**Question 3:Which age group of tourists brought the highest amount of revenue?**


```python
# Total revenues generated as per age group of the tourists
expenditure_by_age_group=df.groupby('age_group')['total_cost'].sum().astype('int').sort_values(ascending=False).reset_index()
plt.figure(figsize=(6,4),dpi=150)
plt.ticklabel_format(style='plain')
sns.barplot(x='age_group',y='total_cost',data=expenditure_by_age_group)
plt.title('Number of tourists and what they came for.')
plt.xticks(rotation=90)
plt.gca().set(frame_on=False)
```

The chat above shows the total amount of revenues that each age group generated to the economy of Tanzania. According to the chart tourists aged 45 and 64 years generated the highest revenue amount while those aged 25 and 44 years followed closely.
